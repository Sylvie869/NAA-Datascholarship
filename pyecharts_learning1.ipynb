3.2 词云图实际例子
我选取了一篇自己用过的论文，一篇有关于创建领域知识图谱的论文，我先利用网页上的pdf转txt进行转换，然后用jieba分词提取关键词词频，最后利用词频来绘制词云图。

生产了一个钻石图案的词云图，从图中可以看出最大的两个单词‘knowledge’和'entity’,就是这篇论文里面的常用专业术语。知识图谱中的知识就是knowledge，实体叫做entity。这个词云图很好地展现了这篇论文的主要内容。

```python
# WordCloud
# 论文文本词云图
import re
import string
import jieba
import jieba.analyse
import html
from numpy import *

txt_path = "data/construction.txt"
with open(txt_path,"rb") as x:
    content = x.read()
    # 正则过滤
    content = re.sub("[{}]+".format(string.punctuation), " ", content.decode("utf-8"))
    # html 转义符实体化
    content = html.unescape(content)
    # 切割
    seg = [i for i in jieba.cut(content, cut_all=True) if i != '']
    # 提取关键词（TF-IDF）
    keywords = jieba.analyse.extract_tags("|".join(seg),
    topK=20, withWeight=True)
    # 分词与关键词提取
    keywords_weight = keywords
    #with open('./keywords0.txt', 'w') as k0:
    #    k0.write(str(keywords0))
    #    k0.close()

    print('完成文章关键字提取！') 
```

```python
# 绘制词云图

## diamond shape
c = (
    WordCloud()
    .add("", keywords_weight, word_size_range=[20, 100], shape=SymbolType.DIAMOND)
    .set_global_opts(title_opts=opts.TitleOpts(title="WordCloud-Article"))
    #.render("wordcloud_diamond.html")
)
c.render_notebook()
```
